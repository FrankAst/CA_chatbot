## Agregrar informacion de contacto

**Objetivo:** traer la informacion de contacto del df_representantes a el df_distri_loc_area_repre. 

Pasos:
* Separar columna de nombres en nombre y apellido en el df_representates.
* Separar columna de nombres en nombre y apellido en df_rtv y df_dtm
* tratar de unir 

```python
# Funcion para generar split de apellidos
def split_name(name):
    
    words = name.split()
    if len(words) == 2:
        # If there are exactly two words, split by the first space
        return words[0], words[1]
    elif len(words) > 2:
        # return first and last word
        return words[0], words[-1]
        #return ' '.join(words[:2]), ' '.join(words[2:]) # If there are more than two words, split by the second space
    else:
        # If there is only one word (just in case), return the word and an empty string
        return words[0], ''
    
```

##### Fuzzy Matching

```python
from fuzzywuzzy import fuzz, process

def fuzzy_merge(df1, df2, key1, key2, threshold=90):
    """
    df1 is the left table to join
    df2 is the right table to join
    key1 is the key column of df1
    key2 is the key column of df2
    threshold is how close the matches should be to return a match, based on Levenshtein distance
    """
    s = df2[key2].tolist()
    
    # Encuentro el match -> m devuelve una serie de pandas: ['(NOMBRE, # VECES QUE LO ENCUENTRA)']
    m = df1[key1].apply(lambda x: process.extractOne(x, s, score_cutoff=threshold))
    
    # Obtengo el nombre de df2 - debo transformar m.
    
    # Step 1: Quito parentesis
    m_clean = m.astype(str).str.strip('()')

    # Step 2: Separo string entre nombre y # de veces que encuentra
    df_split = m_clean.str.split(', ', expand=True)

    # Step 3: Solo me interesa el nombre
    match = df_split[[0]].rename(columns={0: 'Name'})
    
    df1['matches'] = match['Name'] # solo me quedo el nombre
    
    return df1


```

```python
# Creamos merged set
merged_RTV = fuzzy_merge(df_RTV, df_representantes_RTV, 'NOMBRE RTV', 'Nombre', threshold=60)
```

```python
merged_RTV[merged_RTV['matches'].isnull()]
```

```python
# quitamos comillas
merged_RTV['matches'] = merged_RTV['matches'].str.replace("'"," ")

#  agregamos contacto
merged_RTV = merged_RTV.rename(columns={'matches': 'Nombre'})

# Limpiamos
merged_RTV['Nombre'] = merged_RTV['Nombre'].str.strip()
merged_RTV['Nombre'] = merged_RTV['Nombre'].str.replace(r'\s+', ' ', regex=True)
```

```python
# Unimos tabla RTV
merged_RTV = pd.merge(merged_RTV, df_representantes_RTV, on = 'Nombre', how = 'left')

# Quitamos columnas adicionales
merged_RTV.drop(columns= ['nombre_y', 'apellido_y','NOMBRE RTV', 'Distrito', 'Area_Actuacion'], inplace= True)
```
